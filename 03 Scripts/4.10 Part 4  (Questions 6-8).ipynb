```python
"""
File: sample_profiled_updated.pkl
Author:Erica Henry
Date: 3/9/24

#Description: Create bar chart to visualize spending by customer profile.
"""

#Create crosstab of Region and Frequency flag columns to examine differences in ordering habits
cross_spending_profile = pd.crosstab(small['customer_profile'], small['Spending_flag'], dropna = False)
```

```python
cross_spending_profile.to_clipboard()
```

```python
cross_spending_profile
```

```python
#Calculate total count of high and low spenders
cross_spending_profile['Total'] = cross_spending_profile['High spender'] + cross_spending_profile['Low spender']
```

```python
#Calculate proportions of high and low spenders
cross_spending_profile['Proportion high'] =cross_spending_profile['High spender'] / cross_spending_profile['Total'] * 100
```

```python
cross_spending_profile['Proportion low'] = cross_spending_profile['Low spender'] / cross_spending_profile['Total'] * 100
```

```python
#Sort df by frequency
cross_spending_profile = cross_spending_profile.sort_values(by='Total', ascending=False)
```

```python
#View updated crosstab
cross_spending_profile
```

```python
#Create stacked bar chart
cross_spending_profile_bar  = cross_spending_profile[['Proportion high', 'Proportion low']].plot.bar(stacked=True, color=["teal", "darkslategrey", "blue"])
plt.title('Spending by Customer Profile')
plt.xlabel ('Customer Profile')
plt.ylabel('Total Spending')
```

```python
#Save bar chart to Jupyterâ€™s memory 
cross_spending_profile_bar = cross_spending_profile[['Proportion high', 'Proportion low']].plot.bar(stacked=True, color=["teal", "darkslategrey", "blue"])
plt.title('Spending by Customer Profile')
plt.xlabel ('Customer Profile')
plt.ylabel('Total Spending')
```

```python
cross_spending_profile_bar.figure.tight_layout()
```

```python
#Export cross_reg_spend_bar chart in the form of an image .png file
cross_spending_profile_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Spending by Customer Profile.png'))
```
