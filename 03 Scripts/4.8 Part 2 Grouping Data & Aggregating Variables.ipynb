```python
#Question 2 CareerFoundry Task: In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.
##Step 1 and 2  of 3 step procedure
ords_prods_merge.groupby('department_id').agg({'number_of_orders': ['mean']})
```

```python
#Question 3 CareerFoundry Task: Analyze the result. How do the results for the entire dataframe differ from those of the subset? Include your comments in a markdown cell below the executed code.
##Answer: The mean begins with a higher value in the subset and ends with a higher value in the entire df. This process for the entire df gives more extensive information overall.
```

```python
#Question 4 part 1 CareerFoundry Task: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.
##Aggregating ords_prods_merge with transform() and creating a new column 'max_order' containing the maximum frequency of the “number_of_orders” column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number_of_orders'].transform(np.max)
```

```python
#Check results
ords_prods_merge
```

```python
#Check output
ords_prods_merge.head(100)
```

```python
#Question 4  part 2 CareerFoundry Task:
##Deriving Columns with loc() and setting loyalty customer flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Print the frequency of new “loyalty_flag” column 
ords_prods_merge["loyalty_flag"].value_counts(dropna=False)
```

```python
#Confirm the numbers and accuracy of the flag labels
ords_prods_merge[['user_id', 'loyalty_flag', 'number_of_orders']].head(60)
```

```python
#Question 5 CareerFoundry Task: The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.
##Use head() to access multiple columns
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#Question 6 CareerFoundry Task: The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
##Aggregating ords_prods_merge with transform() and creating a new column 'Average Spending' containing the average frequency of the “prices” column
ords_prods_merge['Average Spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check results and output
ords_prods_merge.head()
```

```python
#Deriving Columns with loc() and setting Spending flags
ords_prods_merge.loc[ords_prods_merge['Average Spending'] < 10, 'Spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['Average Spending'] >= 10, 'Spending_flag'] = 'High spender'
```

```python
#Check output
ords_prods_merge.head(50)
```

```python
#Confirm the numbers and accuracy of the flag labels
ords_prods_merge[['user_id', 'Spending_flag', 'prices']].head(60)
```

```python
#Print the frequency of new “Spending_flag” column 
ords_prods_merge["Spending_flag"].value_counts(dropna=False)
```

```python
#Question 7 CareerFoundry Task: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.
##Aggregating ords_prods_merge with transform() and creating a new column 'Median Customer Frequency' containing the median frequency of the “days_since_prior_order” column
ords_prods_merge['Median Customer Frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Check results
ords_prods_merge.head(60)
```

```python
#Deriving Columns with loc() and setting customer frequency flags
ords_prods_merge.loc[ords_prods_merge['Median Customer Frequency'] <20, 'Frequency flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['Median Customer Frequency'] >10) & (ords_prods_merge['Median Customer Frequency'] >= 20), 'Frequency flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'Frequency flag'] = 'Frequent customer'
```

```python
#Check output
ords_prods_merge.head(50)
```

```python
#Confirm the numbers and accuracy of the flag labels
ords_prods_merge[['user_id','Frequency flag']].head(60)
```

```python
#Print the frequency of new “Frequency flag” column 
ords_prods_merge["Frequency flag"].value_counts(dropna=False)
```
