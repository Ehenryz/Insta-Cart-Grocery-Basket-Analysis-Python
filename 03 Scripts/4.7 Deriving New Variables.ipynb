```python
#Create a subset of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
#Define a function, create a flag to sort products into price ranges in subset (If-Statements with User-Defined Functions)
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function to subset
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Print frequency of new column price_range in subset
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
#If-Statements with the loc() Function, create conditions to sort products into price ranges in ords_prods_merge
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Print frequency of new column price_range_loc in ords_prods_merge
ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

```python
#Print frequency of column orders_day_of_week in ords_prods_merge
ords_prods_merge["orders_day_of_week"].value_counts(dropna = False)
```

```python
#If-Statements with For-Loops practice, create for-loop in ords_prods_merge to summarize how busy each day of the week is
result = []
for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Print result list
result 
```

```python
#Create new column busiest_day in ords_prods_merge df for results list
ords_prods_merge["busiest_day"] = result
```

```python
#Print frequency of new column busiest_day
ords_prods_merge["busiest_day"].value_counts(dropna = False)
```

```python
#Question 2 CareerFoundry Task: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.
result = []
for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1: 
    result.append("2nd Busiest day")
  elif value == 3:
    result.append("2nd Least busy day")
  elif value == 4:
    result.append("Least busy day")
  else:
    result.append("Regularly busy day")
```

```python
#Question 3 CareerFoundry Task: Check the values of this new column for accuracy. Note any observations in markdown format.
result
```

```python
#Create new column in ords_prods_merge df for results list
ords_prods_merge["busiest_days"] = result
```

```python
#Print frequency of new column busiest_days
ords_prods_merge["busiest_days"].value_counts(dropna = False)
```

```python
#Print frequency of column order_hour_of_day
ords_prods_merge["order_hour_of_day"].value_counts(dropna = False)
```

```python
#Using if-statements with for-loops to determine busiest hours of the day
result = []
for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10:
    result.append("Most orders")
  elif value == 3: 
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
#Print result list
result
```

```python
#Create new column busiest_period_of_day in ords_prods_merge df for results list
ords_prods_merge["busiest_period_of_day"] = result
```
